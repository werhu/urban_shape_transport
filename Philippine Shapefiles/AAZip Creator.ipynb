```python
import zipfile, os
import numpy as np
import pandas as pd
```

```python
provinces = pd.read_csv('AARegions.csv')
```

```python
for i in provinces['Regions']:
    zip_file = zipfile.ZipFile('Zip Files/'+i+'.zip', 'w')

    zip_file.write(i+'.cpg')
    zip_file.write(i+'.dbf')
    zip_file.write(i+'.shp')
    zip_file.write(i+'.shx')
    zip_file.write(i+'.prj')
    
    zip_file.close()
```
